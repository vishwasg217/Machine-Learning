# Gradient Descent

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

def func(x):
    return 3*x+2

data = pd.read_csv('study.csv')
x = data.iloc[:, 0]
y = data.iloc[:, 1]



# x = np.arange(50)
# y = np.array(list(map(func, x)))
# fx = 3*x+2
plt.scatter(x, y)
#plt.plot(x, y, '-r')
#plt.xlim([0, 13000])
#plt.ylim([0, 11000000])
plt.show()
```

```python

m = 0
b = 0
m_list = []
b_list = []
cost_list = []
y_hat = 0
def gradient_descent(x, y):
    curr_m = curr_b = 0
    epochs = 100
    learning_rate = 0.01
    n = len(x)
    global cost_list, b_list, m_list

    for i in range(epochs):
        y_hat = curr_m*x + curr_b
        cost = (1/n) * sum((y-y_hat)**2)
        cost_list.append(cost)
        der_m = -(2/n) * sum(x*(y-y_hat))
        der_b = -(2/n) * sum(y-y_hat)
        curr_m -= learning_rate*der_m
        m_list.append(curr_m)
        curr_b -= learning_rate*der_b
        b_list.append(curr_b)
        print("cost: ",cost, "slope: ", curr_m, "intercept: ", curr_b)

    global m, b
    m = curr_m
    b = curr_b
    

gradient_descent(x, y)
#print(m, b)
#print("cost: ", cost_list, "slope: ", m_list, "intercept: ", b_list)
```

```python
plt.scatter(x, y)
#plt.plot(x, y, '-r')
y_pred = m*x + b
plt.plot(x, y_pred, '-g')

```

```python
plt.grid()
plt.scatter(x, y)
y_pred = m*x + b
plt.plot(x, y_pred, '-g')

n = len(x)

def mse(x, y):
    cost = (1/n) * sum((y-y_pred)**2)   
    print("cost: ", cost)

mse(x, y)

distance = abs(y-y_pred)

for i in range(len(distance)):
    plt.plot([x[i], x[i]], [y[i], y_pred[i]], 'k--')


```
